```python
import pandas as pd
```

```python
median_sale_price = pd.read_csv("Metro_median_sale_price_uc_sfrcondo_month.csv")
median_sale_price_san_francisco = median_sale_price[median_sale_price["RegionName"] == "San Francisco, CA"]
median_sale_price_san_francisco 
```

```python
median_sale_price_san_francisco_vertical = median_sale_price_san_francisco.transpose()
```

```python
median_sale_price_san_francisco_vertical.columns = ['Median_Sale_Price']
```

```python
median_sale_price_san_francisco_vertical.index = pd.to_datetime(median_sale_price_san_francisco_vertical.index)
median_sale_price_san_francisco_vertical.head()
```
