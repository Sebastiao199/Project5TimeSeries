```python
def scraping_idealista_rent2():
    url = 'file:///C:/Users/sebas/Desktop/WildCodeSchool/Project5/rent_idealista_squaredmeter_price.html'
    driver.get(url)
    soup_selenium = BeautifulSoup(driver.page_source, 'html.parser')
    table = soup_selenium.find_all('tr', {'class': 'table__row'})
    data_sortable = list(map(lambda item: item.text, table))
    # time.sleep(1)
    return data_sortable

result_rent2 = scraping_idealista_rent2()
result_rent2
```

```python
price_only_rent = list(map(lambda item: item[4:], year_and_price_rent))
price_only_rent
```

```python
price_only_rent_final = list(map(lambda item: item.replace(",", "."), price_only_rent))
price_only_rent_final
```
